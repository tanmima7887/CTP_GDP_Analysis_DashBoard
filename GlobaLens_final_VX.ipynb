```python
#import packages
import base64
from PIL import Image
from PIL import Image, ImageDraw, ImageFont
from io import BytesIO

import pandas as pd
import numpy as np
import panel as pn
pn.extension('tabulator')

import hvplot.pandas
import plotly.graph_objects as go

import dash
from dash import dcc, html
from dash.dependencies import Input, Output, State
import pandas as pd
import plotly.express as px
import numpy as np
import seaborn as sns
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
import math
import statsmodels.api as sm
import matplotlib.ticker as mtick

pd.options.display.float_format = '{:.0f}'.format
```

```python
df = pd.read_csv('data/fertility2.csv')
df2 = pd.read_csv('data/Countries.csv')
```

```python
# restructred 

df_melt = pd.melt(df, id_vars=['Series Name', 'Series Code', 'Country Name', 'Country Code'], 
                  var_name='Year', value_name='Value')
df_melt['Year'] = df_melt['Year'].str.extract('(\d{4})')[0]
df_restructured = df_melt.pivot_table(index=['Country Name', 'Country Code', 'Year'], 
                                      columns='Series Name', 
                                      values='Value', 
                                      aggfunc='first').reset_index()
df_restructured.head()
```

```python
app = dash.Dash(__name__)
min_year = df_restructured['Year'].min()
max_year = df_restructured['Year'].max()
image_index = 0
# Slides
#pil_image1 = Image.open("images\Slide1.JPG")
image1 = Image.open("images\Slide1.JPG")
image2 = Image.open("images\Slide2.JPG")
image3 = Image.open("images\Slide3.JPG")
image4= Image.open("images\Slide4.JPG")
image5 = Image.open("images\Slide5.JPG")
image6 = Image.open("images\Slide6.JPG")
image7 = Image.open("images\Slide7.JPG")
image8 = Image.open("images\Slide8.JPG")
image9 = Image.open("images\Slide9.JPG")
image10 = Image.open("images\Slide10.JPG")
image11 = Image.open("images\Slide11.JPG")
image12 = Image.open("images\Slide12.JPG")

# Encode images to base64 for displaying in Dash
encoded_image1 = base64.b64encode(open("images\Slide1.JPG", "rb").read()).decode('ascii')
encoded_image2 = base64.b64encode(open("images\Slide2.JPG", "rb").read()).decode('ascii')
encoded_image3 = base64.b64encode(open("images\Slide3.JPG", "rb").read()).decode('ascii')
encoded_image4 = base64.b64encode(open("images\Slide4.JPG", "rb").read()).decode('ascii')
encoded_image5 = base64.b64encode(open("images\Slide5.JPG", "rb").read()).decode('ascii')
encoded_image6 = base64.b64encode(open("images\Slide6.JPG", "rb").read()).decode('ascii')
encoded_image7 = base64.b64encode(open("images\Slide7.JPG", "rb").read()).decode('ascii')
encoded_image8 = base64.b64encode(open("images\Slide8.JPG", "rb").read()).decode('ascii')
encoded_image9 = base64.b64encode(open("images\Slide9.JPG", "rb").read()).decode('ascii')
encoded_image10 = base64.b64encode(open("images\Slide10.JPG", "rb").read()).decode('ascii')
encoded_image11 = base64.b64encode(open("images\Slide11.JPG", "rb").read()).decode('ascii')
encoded_image12 = base64.b64encode(open("images\Slide12.JPG", "rb").read()).decode('ascii')


# Layout of the app
custom_styles = {
    'backgroundColor': '#f4f4f4',  # Background color
    'fontFamily': 'Arial, sans-serif',  # Font family
    'fontSize': 16,  # Font size
    'textAlign': 'center',  # Text alignment
    'padding': '20px'  # Padding
}
app.layout = html.Div(children=[
    html.H1("World GDP Analytics Dashboard", style={'textAlign': 'center', 'color': '#000000'}),

    html.Div([
        html.H2("Global Economic Factors", style=custom_styles),

        html.Label("Select Y-axis"),
        dcc.Dropdown(
            id='y-axis-dropdown-1',
            options=[{'label': col, 'value': col} for col in df2.columns[3:]],
            value=df2.columns[3],
            multi=False
        ),

        html.Label("Select Countries"),
        dcc.Dropdown(
            id='country-dropdown-1',
            options=[{'label': country, 'value': country} for country in df2['Country Name'].unique()],
            value=df2['Country Name'].unique(),
            multi=True
        ),

        dcc.Graph(id='line-plot-1')
    ]),

    html.Div([
        html.H2("Global Population Variable Data", style=custom_styles),

        html.Label("Select X-axis"),
        dcc.Dropdown(
            id='x-axis-dropdown',
            options=[
                {'label': 'Year', 'value': 'year'},
                * [{'label': col, 'value': col} for col in df_restructured.columns[3:]],
            ],
            value='year',
            multi=False
        ),

        html.Label("Select Y-axis"),
        dcc.Dropdown(
            id='y-axis-dropdown-2',
            options=[{'label': col, 'value': col} for col in df_restructured.columns[3:]],
            value=df_restructured.columns[3],
            multi=False
        ),

        html.Label("Select Countries"),
        dcc.Dropdown(
            id='country-dropdown-2',
            options=[{'label': country, 'value': country} for country in df_restructured['Country Name'].unique()],
            value=df_restructured['Country Name'].unique(),
            multi=True
        ),

        dcc.Graph(id='line-plot-2'),
        #image
        html.H3(style={'margin-bottom': '100px'}),
        #html.Img(src=pil_image1, style={'width': '80%', 'margin': 'auto', 'display': 'block'})
    ]),
    html.Div([
    html.Img(id='image-display', src=f'data:image/png;base64,{encoded_image1}', style={'width': '80%', 'margin': 'auto', 'display': 'block'}),
    html.Button('Next Image', id='button-next-image', n_clicks=0),
])

])


@app.callback(
    Output('line-plot-1', 'figure'),
    [Input('y-axis-dropdown-1', 'value'),
     Input('country-dropdown-1', 'value')]
)
def update_graph_1(selected_y, selected_countries):
    filtered_df = df2[(df2['Country Name'].isin(selected_countries))]

    # Create line plot
    fig = px.line(
        filtered_df,
        x='Year',
        y=selected_y,
        color='Country Name',
        labels={'Year': 'Year', 'value': 'Selected Variables'},
        title='Selected Variables over Time'
    )

    return fig

@app.callback(
    Output('line-plot-2', 'figure'),
    [Input('x-axis-dropdown', 'value'),
     Input('y-axis-dropdown-2', 'value'),
     Input('country-dropdown-2', 'value')]
)
def update_graph_2(selected_x, selected_y, selected_countries):
    filtered_df = df_restructured[df_restructured['Country Name'].isin(selected_countries)]

    # Create line plot
    fig = px.line(
        filtered_df,
        x=selected_x if selected_x != 'year' else 'Year',
        y=selected_y,
        color='Country Name',
        labels={'value': 'Selected Variables'},
        title='Selected Variables over Time'
    )

    return fig

@app.callback(
    [Output('image-display', 'src'),
     Output('button-next-image', 'n_clicks')],
    Input('button-next-image', 'n_clicks'),
    prevent_initial_call=True
)
# Every click change picture/ slides
def update_image_and_count(n_clicks):
    global image_index

    image_index = (image_index + 1) % 12 

    if image_index == 0:
        return f'data:image/png;base64,{encoded_image1}', n_clicks + 1
    elif image_index == 1:
        return f'data:image/png;base64,{encoded_image2}', n_clicks + 1
    elif image_index == 2:
        return f'data:image/png;base64,{encoded_image3}', n_clicks + 1
    elif image_index == 3:
        return f'data:image/png;base64,{encoded_image4}', n_clicks + 1
    elif image_index == 4:
        return f'data:image/png;base64,{encoded_image5}', n_clicks + 1
    elif image_index == 5:
        return f'data:image/png;base64,{encoded_image6}', n_clicks + 1
    elif image_index == 6:
        return f'data:image/png;base64,{encoded_image7}', n_clicks + 1
    elif image_index == 7:
        return f'data:image/png;base64,{encoded_image8}', n_clicks + 1
    elif image_index == 8:
        return f'data:image/png;base64,{encoded_image9}', n_clicks + 1
    elif image_index == 9:
        return f'data:image/png;base64,{encoded_image10}', n_clicks + 1
    elif image_index == 10:
        return f'data:image/png;base64,{encoded_image11}', n_clicks + 1
    else:
        return f'data:image/png;base64,{encoded_image12}', n_clicks + 1

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True,  port=7090)
    #http://127.0.0.1:7090/
```
